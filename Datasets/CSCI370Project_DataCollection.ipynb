Code for 1 video:

```python
import googleapiclient.discovery
import pandas as pd

# ── Config ──────────────────────────────────────────────────────────
API_KEY = "AIzaSyCw659cbO7KRtRIeCOLebEA75flOw-HMEc"
VIDEO_ID = "02v6NIsrsM"


# ── Build client ─────────────────────────────────────────────────────
youtube = googleapiclient.discovery.build("youtube", "v3", developerKey=API_KEY)


# ── Helper: fetch all comments for a video ───────────────────────────
def get_comments(video_id: str) -> pd.DataFrame:
    rows = []
    request = youtube.commentThreads().list(
        part="snippet", videoId=video_id, maxResults=100
    )

    while request:
        response = request.execute()
        for item in response["items"]:
            snippet = item["snippet"]["topLevelComment"]["snippet"]
            rows.append({
                "author":     snippet["authorDisplayName"],
                "updated_at": snippet["publishedAt"],
                "like_count": snippet["likeCount"],
                "text":       snippet["textOriginal"],
                "video_id":   snippet["videoId"],
                "public":     item["snippet"]["isPublic"],
            })
        request = youtube.commentThreads().list_next(request, response)

    return pd.DataFrame(rows)


# ── Single video ──────────────────────────────────────────────────────
df = get_comments(VIDEO_ID)
print(df.info())
print(df.head(10))
print("\nTop 10 by likes:")
print(df.sort_values("like_count", ascending=False).head(10))
```

```python
import googleapiclient.discovery
import pandas as pd

# ── Config ──────────────────────────────────────────────────────────
API_KEY = "AIzaSyCw659cbO7KRtRIeCOLebEA75flOw-HMEc"

# ── Build client ─────────────────────────────────────────────────────
youtube = googleapiclient.discovery.build("youtube", "v3", developerKey=API_KEY)


# ── Helper: fetch all comments for a video ───────────────────────────
def get_comments(video_id: str) -> pd.DataFrame:
    rows = []
    request = youtube.commentThreads().list(
        part="snippet", videoId=video_id, maxResults=100
    )

    while request:
        response = request.execute()
        for item in response["items"]:
            snippet = item["snippet"]["topLevelComment"]["snippet"]
            rows.append({
                "author":     snippet["authorDisplayName"],
                "updated_at": snippet["publishedAt"],
                "like_count": snippet["likeCount"],
                "text":       snippet["textOriginal"],
                "video_id":   snippet["videoId"],
                "public":     item["snippet"]["isPublic"],
            })
        request = youtube.commentThreads().list_next(request, response)

    return pd.DataFrame(rows)


# ── Fetch all videos ──────────────────────────────────────────────────
all_comments = pd.concat([get_comments(vid) for vid in VIDEO_IDS], ignore_index=True)

print(all_comments["video_id"].value_counts())
print(all_comments.info())
print(all_comments.head(10))
print("\nTop 10 by likes:")
print(all_comments.sort_values("like_count", ascending=False).head(10))
```
